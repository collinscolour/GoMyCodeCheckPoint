```python
  #Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import metrics
from scipy import stats
from sklearn.preprocessing import LabelEncoder
```

```python
#import csv file
df = pd.read_csv('African_crises_dataset.csv')
df.info()
```

```python
from ydata_profiling import ProfileReport
# Generate the profiling report
profile = ProfileReport(df)

# Save the report to HTML
profile.to_file('report.html')
```

```python
# Clean the data
# Check for missing values
df.isnull().sum()
```

```python
df.info()
```

```python
#encoding categorical features using LabelEncoder
categorical_features = df.select_dtypes(include=['object', 'category']).columns
for feature in categorical_features:
  encoder = LabelEncoder()
  df[feature] = encoder.fit_transform(df[feature])
```

```python
df.describe()
```

```python
df1 = df[(np.abs(stats.zscore(df['inflation_annual_cpi']))<3) & (np.abs(stats.zscore(df['exch_usd']))<3)]

```

```python
df1
```

```python
# Save the cleaned data to a new CSV file
df1.to_csv('cleaned_data.csv', index=False)
```

```python
df1.isnull().sum()
```

```python
y=df1["systemic_crisis"]
X=df1.drop("systemic_crisis", axis=1)
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=30) #splitting data with test size of 35%
model=LogisticRegression(max_iter=500)   #build Logistic Regression  model
model.fit(x_train,y_train)  #fitting the training data
predicted=model.predict(x_test) #testing our model’s performance
accuracy= accuracy_score(y_test,predicted)
precision= precision_score(y_test,predicted)
recall= recall_score(y_test,predicted)
f = f1_score(y_test,predicted)
print( 'Accuracy',accuracy)
print( 'Precision', precision)
print( 'Racall', recall)
print( 'Fi-score', f)
```
